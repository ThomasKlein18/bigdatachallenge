```python
def clusterfilter(data, windowsize,num):
    res = np.zeros(data.shape)
    half = windowsize//2
    mean = np.mean(data)
    std = np.std(data)
    for i in range(half,len(data)-half):
        count = 0
        for val in data[i-half:i+half]:
            if val > mean+std or val < mean-std:
                count += 1
                if count > num:
                    res[i] = 1
                    break
    return res
```
